```python
def VarChecker(list1, list2):
    """
    
    Cette fonction sert à vérifier si la variable sélectionnée est dans la liste des variables.
    
    Args:
        list1 : liste des variables du jeu de données
        list2 : liste des variables sélectionnées
        
    Returns:
        Variable(s) en commun dans les deux listes
        
    """
    section = list(set(list1).intersection(set(list2)))
    return section
```

```python
VarChecker(list_ind, vars_internet)
```

```python
vars_internet = VarChecker(list_ind, vars_internet)
```

```python
msno.matrix(df_country_ind_sort[df_country_ind_sort.indicator_name.isin(vars_internet)])
plt.show()
```
