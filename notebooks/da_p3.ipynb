### Import et préparation du dataset

```python
veg = pd.read_csv('FAOSTAT_2010_2013_BA_veg.csv')
ani = pd.read_csv('FAOSTAT_2010_2013_BA_ani.csv')
```

```python
# Ajout de la variable origin
ani["origin"] = "animal"
veg["origin"] = "vegetal"

# On regroupe veg et ani en un unique dataframe, via une union
temp = ani.append(veg)
```

```python
temp
```

```python
# On renomme les colonnes de temp
temp.columns = ["xx","xx2","country_code","country",'xx3','element'
    ,'item_code','item','xx4',"year","unit","value",'xx5','xx6'
    ,'origin']
```

```python
# Transformation de temp en table pivot
data = temp.pivot_table(
    index=["country_code","country","item_code","item","year","origin"],
    columns = "element", values="value", aggfunc=sum)
```

```python
# Aperçu des données
data
```

```python
# Changement de noms des colonnes
data.columns = ['aliments_pour_animaux, milliers_de_tonnes', 'autres_utilisations, milliers_de_tonnes',
'dispo_alim, kcal_personne_jour', 'dispo_alimen_quanti, kg_personne_an', 'dispo_mat_grasse, g_personne_jour', 
'dispo_prot, g_personne_jour', 'dispo_int, milliers_de_tonnes', 'export, milliers_de_tonnes', 
'import, milliers_de_tonnes', 'nourriture, milliers_de_tonnes', 'pertes, milliers_de_tonnes', 
'production, milliers_de_tonnes', 'semences, milliers_de_tonnes', 'traitement, milliers_de_tonnes', 
'var_stock, milliers_de_tonnes']
```

```python
data = data.reset_index()
```

### Merge des deux Data Frame

```python
data2 = pd.merge(pop2, data, on=['country', 'year'])
data2
```

### Calculs

```python
# on multiplie les disponiblités unitaires par le nombre d'habitants pour chaque produit
data2['total_dispo_kcal'] = data2['population, 1000 personnes'] * data2['dispo_alim, kcal_personne_jour'] * 1000 *365
data2['total_dispo_prot_kg'] = data2['population, 1000 personnes'] * data2['dispo_prot, g_personne_jour'] * 365
```

```python
data2.head()
```

# Q4

```python
# Création de la colonne Ratio énergétiqu
data2['ratio énergétique, Kcal/kg'] = (data2['dispo_alim, kcal_personne_jour'] * data2['population, 1000 personnes'] * 1000 * 365) / (data2['nourriture, milliers_de_tonnes'] * 1000 * 1000)
```

```python
data2.loc[data2['item'] == 'Oeufs']
# la valeur calorique d'un oeuf est de 155 calories pour 100g (Wikipedia), les données semblent cohérentes
```

```python
# Création de la colonne Ratio protéine
data2['ratio protéine'] = (data2['dispo_prot, g_personne_jour'] * 365 * data2['population, 1000 personnes'] * 1000) / (data2['nourriture, milliers_de_tonnes'] * 1000 * 1000 * 1000)
data2.head()
```

Remplacement des 'inf' et des '0' par des NaN pour éviter la division par 0

```python
data2['ratio énergétique, Kcal/kg'] = data2['ratio énergétique, Kcal/kg'].replace(np.inf, np.nan).replace(0, np.nan)
```

```python
data2['ratio protéine'] = data2['ratio protéine'].replace(np.inf, np.nan).replace(0, np.nan)
```
